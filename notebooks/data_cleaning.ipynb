```python
import pandas as pd
import numpy as np

# Load the dataset
# The '../' goes up one level from 'notebooks' to the main project folder, then into 'data'
# low_memory=False helps prevent a warning about mixed data types in columns.
df = pd.read_csv('../data/loan.csv', low_memory=False)
# Display the first 5 rows to see what the data looks like
df.head()
```

```python
# Get a summary of the DataFrame including data types and non-null counts
df.info(verbose=True, show_counts=True)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean()

# Get the list of columns to drop (those with more than 40% missing data)
cols_to_drop = missing_percentage[missing_percentage > 0.4].index

# Drop these columns from the DataFrame
df.drop(columns=cols_to_drop, inplace=True)

# Check the new shape of our DataFrame
print(f"Original number of columns: {len(missing_percentage)}")
print(f"Number of columns after dropping: {df.shape[1]}")
```

```python
df.info(verbose=True, show_counts=True)
```

```python
# Clean 'int_rate': remove '%' and convert to float
# df['int_rate'] = df['int_rate'].str.replace('%', '').astype(float)

# See how many rows have a missing term
print("Number of missing terms:", df['term'].isnull().sum())
print("Total rows:", len(df))

# If the number is small, drop them
# df.dropna(subset=['term'], inplace=True)

# Now this conversion will work without error
# df['term'] = df['term'].str.replace(' months', '').astype(int)

# Clean 'emp_length': extract the numbers. We'll fill missing values with 0.
# This uses a regular expression to find the digits.
# df['emp_length'] = df['emp_length'].str.extract('(\d+)').astype(float)
df['emp_length'] = df['emp_length'].fillna(0) # Assume missing means 0 years experience or unemployed
```

```python
df.info(verbose=True, show_counts=True)
```

```python
# See the different values in the 'loan_status' column
print(df['loan_status'].value_counts())

# Define what constitutes a "bad loan"
bad_loan_statuses = ['Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off', 'Late (31-120 days)']

# Create our target variable: 1 for a bad loan, 0 for a good loan
df['is_bad_loan'] = np.where(df['loan_status'].isin(bad_loan_statuses), 1, 0)

# See the distribution of our new target variable
print(df['is_bad_loan'].value_counts(normalize=True))
```

```python
# Save the cleaned DataFrame to a new file in the data directory
df.to_csv('../data/processed_data.csv', index=False)
```
